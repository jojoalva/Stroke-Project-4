# Cleaned dataset algorithms

#### This dataset was cleaned, by converting the age to its nearest integer, and removed Gender "other" which was 1 row of data, and we also removed 'smoking_status' from the original data. The age didn't make sense in decimal points as some candidates were less than 1 year old. Smoking status was dropped because we originally ran a features importance on the dataset and decided to remove it.

```python
# Import the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import balanced_accuracy_score, accuracy_score, confusion_matrix, classification_report
```

## 1. Pre-processing data

```python
# Read the optimised & cleaned CSV file from the Resources folder into a Pandas DataFrame
df = pd.read_csv(
    Path('stroke_cleaned_v2.csv')
)

# Review the DataFrame
df.head()
```

```python
# Get dummies to convert all data to numerical data type
df = pd.get_dummies(df)
df.head()
```

## 2. Splitting the data into training and test data

```python
# Separate the data into labels and features

# Separate the y variable, the labels from the 'stroke' column
y = df['stroke']

# Separate the X variable, the features by dropping the 'stroke' column
X = df.drop(columns='stroke')
```

```python
# Review the y variable Series
y.ravel()
```

```python
# Drop ID column
X = X.drop(columns='id')
# Convert NaN values in BMI to mean BMI
mean_bmi = X['bmi'].mean()
X['bmi'] = X['bmi'].fillna(mean_bmi)
#checking after filling the null values with mean
print("Number of missing values in 'bmi' after imputation:", X['bmi'].isnull().sum())
```

```python
# Review the X variable DataFrame
X
```

```python
# Check the balance of our target values
y.value_counts()
```

## 3. Optimisation 1: oversampling data

```python
# Split the data using train_test_split
# Assign a random_state of 78 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# # Assign a random_state parameter of 1 to the model
ros = RandomOverSampler(sampling_strategy='minority', random_state=78)

# Fit the original training data to the random_oversampler model
x_res, y_res = ros.fit_resample(X_train, y_train)
```

```python
# Count the distinct values of the resampled labels data
y_res.value_counts()
```

### Scaling Data

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Using Random Forest Classifier Algorithm

```python
# Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

#### False negative is 67, this is a high risk group. They have been incorrectly identified as 'no-stroke' when in reality the result is 'stroke'. This has resulted in poor precision and recall for 'stroke' prediction, so this needs to be improved. We can improve this but first we need to look at what features in the data is most important when it comes to training the algorithm.

```python
# Visualize the features by importance
importances_df = pd.DataFrame(sorted(zip(rf_model.feature_importances_, X.columns), reverse=True))
importances_df.set_index(importances_df[1], inplace=True)
importances_df.drop(columns=1, inplace=True)
importances_df.rename(columns={0: 'Feature Importances'}, inplace=True)
importances_sorted = importances_df.sort_values(by='Feature Importances')
importances_sorted.plot(kind='barh', color='lightgreen', title= 'Features Importances', legend=False)
```

## 6. Optimisation 4: Outliers dropped from original dataset to see if recall can be further improved

```python
# Drop ID column
df2 = df2.drop(columns='id')
# Convert NaN values in BMI to mean BMI
mean_bmi = df2['bmi'].mean()
df2['bmi'] = df2['bmi'].fillna(mean_bmi)
#checking after filling the null values with mean
print("Number of missing values in 'bmi' after imputation:", df2['bmi'].isnull().sum())
df2
```

```python
nums = list(df2.select_dtypes(include=['int64','float64']))

plt.figure(figsize=(15, 7))
for i in range(0, len(nums)):
    plt.subplot(2, 3, i+1)
    sns.boxplot(y=df2[nums[i]],color='orange',orient='v')
    plt.tight_layout()
```

## 7. Optimisation 5: Balancing no-stroke to match stroke numbers, and then oversampling with Random Forest Classifier

```python
# Read the optimised & cleaned CSV file from the Resources folder into a Pandas DataFrame
df3 = pd.read_csv(
    Path('stroke_cleaned_v2.csv')
)

# Review the DataFrame
df3.head()
```

```python
df3['stroke'].value_counts()
```

```python
# Keep 249 data points where stroke is 0
keep = list(df3[df3.stroke == 0].index[:249])

# Keep all rows where stroke is 1
keep2 = list(df3[df3.stroke == 1].index)

# Combine the indices to create the final filter
filter = keep + keep2
```

```python
df3 = df3.loc[filter]
df3
```

```python
## Splitting the data into training and test data

# Separate the data into labels and features

# Separate the y variable, the labels from the 'stroke' column
y = df3['stroke']

# Separate the X variable, the features by dropping the 'stroke' column
X = df3.drop(columns='stroke')

# Review the y variable Series
y.ravel()

# Convert NaN values in BMI to mean BMI
mean_bmi = X['bmi'].mean()
X['bmi'] = X['bmi'].fillna(mean_bmi)
#checking after filling the null values with mean
print("Number of missing values in 'bmi' after imputation:", X['bmi'].isnull().sum())
# Get Dummies on X
X = pd.get_dummies(X)

# Review the X variable DataFrame
X.head()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Split the data using train_test_split
# Assign a random_state of 78 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# # Assign a random_state parameter of 1 to the model
ros = RandomOverSampler(sampling_strategy='minority', random_state=78)

# Fit the original training data to the random_oversampler model
x_res, y_res = ros.fit_resample(X_train, y_train)

# Count the distinct values of the resampled labels data
y_res.value_counts()
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()

# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)

# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
